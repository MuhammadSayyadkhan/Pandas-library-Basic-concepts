### Dropna && fillna :

```python
import pandas as pd
var = pd.read_csv('E:\\python_machine_documments\\CSV_File.csv')
var
```

### dropna():
What it does: Removes (drops) rows or columns that have missing values (NaN).

#### Remove rows with missing values:
 If any row has a missing value, it will be removed. use dropna().

```python
var1 = pd.read_csv('E:\\python_machine_documments\\CSV_File.csv')
var1.dropna() # Remove any row with value null
```

#### Remove columns with missing values:
if any columns has a missing vlaue, it will be removed. use dropna(axis=1)

```python
var2 = pd.read_csv('E:\\python_machine_documments\\CSV_File.csv')
var2.dropna(axis=1) # mong sara kam columns k om missing value v hagha columns ba full delate she.
```

### how():
how function mong de dapara use kao, kala che mong sara mukamal row khali v null v ao da mong display kawal na ghwaro no mun how use kao.
Or:
ka mong sara somra null row hagha ba ghondy delate she.
#### Note:
any che mong use ko no mong sara ba che somra khali rows v hagha ba delate she.
all che use ko no che mong sara km particular rows v hagha ba delate she.

```python
var3 = pd.read_csv('E:\\python_machine_documments\\CSV_File.csv')
var3.dropna(how="any") # somra rwos om v hagha delate shwy
```

```python
var = pd.read_csv('E:\\python_machine_documments\\CSV_File.csv')
var.dropna(how="all") # mong sara ba particular rows delate she
```

### subset():
subset mong de depara use kao , kala che mong 1 particular columns na null value delate kao, remove kao.

```python
var = pd.read_csv('E:\\python_machine_documments\\CSV_File.csv')
var.dropna(subset=["Salary"]) # mong sara columns k columns k somra null values v hagha delate shwy.
```

### Inplace():
inplace che mong use ko no mong sara somra om null value de hagha va delate she, remove ba she.

```python
var = pd.read_csv('E:\\python_machine_documments\\CSV_File.csv')
var.dropna(inplace=True)
var
```

### tresh():
tresh function k wrta mong y che single value or double value null delate ka.

```python
var = pd.read_csv('E:\\python_machine_documments\\CSV_File.csv')
var.dropna(thresh=2)
var
```

### fillna():
fillna function mung de dapara use kao, kala che mong Null values pa xai sa name k jaga python.

```python
var = pd.read_csv('E:\\python_machine_documments\\CSV_File.csv')
var.fillna("Python") # km xai k null value wa halta python show kigy ba
```

ka har columns la particular value name wrkao no bia ba mong dictionary use kao.

```python
var = pd.read_csv('E:\\python_machine_documments\\CSV_File.csv')
var.fillna({"Age":"C++", "Salary":"Java", "class":"Javascript"}) # mong sara particular name initialize shwal.
```

### limit():
limit mong de para use kao, kala che mong limit pory data fill kao,

```python
var = pd.read_csv('E:\\python_machine_documments\\CSV_File.csv')
var.fillna("python",limit=2)
```
