```python
import pandas as pd
```

```python
var = {"Days":[1,2,3,4,5,6],"Eng":[13,14,18,17,13,12],"Maths":[15,17,19,16,15,19]}
var_new = pd.DataFrame(var)
var_new
pd.melt(var_new)   # this is simple convert to long format
```

```python
var = {"Days":[1,2,3,4,5,6],"Eng":[13,14,18,17,13,12],"Maths":[15,17,19,16,15,19]}
var_new = pd.DataFrame(var)
pd.melt(var_new,id_vars=["Days"])
```
